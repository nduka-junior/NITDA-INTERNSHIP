```python
from google.colab import files
import pandas as pd
uploaded  = files.upload()
```

```python
merged_data= pd.DataFrame()
for file_name in uploaded:

    data = pd.read_csv(file_name)
    merged_data = pd.concat([merged_data, data,], ignore_index=True)



merged_data.head()
removeNaN = merged_data.dropna(subset=['Order ID'])
removeNaN.head()
```

```python
sort_data = removeNaN.sort_values(by="Order Date", ascending=True)
sort_data.head()
```

```python
sort_data.head()
sort_data = sort_data[sort_data['Order Date'] != 'Order Date']
sort_data.head()
sort_data['Order Date'] = pd.to_datetime(sort_data['Order Date'], format='%m/%d/%y %H:%M')
# sort_data['Price Each'] = sort_data['Price Each'].astype(int)
sort_data['Price Each'] = pd.to_numeric(sort_data['Price Each'], errors='coerce')
sort_data['Price Each'] = sort_data['Price Each'].astype(int)

# Calculate total revenue for each row
sort_data['Total Revenue'] = sort_data['Quantity Ordered'] * sort_data['Price Each']

# Resample by month and calculate sum
monthly_summary = sort_data.resample('M', on='Order Date').sum()
monthly_summary.head()
```

```python

highest_revenue = monthly_summary["Price Each"].max()

month_with_highest_revenue =monthly_summary.idxmax()
month_with_highest_revenue = month_with_highest_revenue["Price Each"]

# Create a Timestamp object
timestamp = pd.Timestamp(month_with_highest_revenue)

# Convert the Timestamp to the month name in full text
month_with_highest_revenue = timestamp.strftime('%B')

print(month_with_highest_revenue)

print("The month with the highest revenue is " +  month_with_highest_revenue + " with a revnue of " + str(highest_revenue) )
```

```python
# CITY WITH MOST  SOLD PRODUCT
import matplotlib.pyplot as plt
cities = sort_data
cities.head()

cities["City"] = cities["Purchase Address"].apply(lambda x: x.split(", ")[1])



# group_cities = cities.groupby("City")
cities['Quantity Ordered'] = pd.to_numeric(cities['Quantity Ordered'], errors='coerce')
cities['Price Each'] = cities['Price Each'].astype(int)


group_cities = cities.groupby('City')['Quantity Ordered'].sum()


city_with_most_sold = group_cities.max()
city_with_most_sold = group_cities[group_cities == city_with_most_sold ].index.tolist()

print("City with most sold product is " + " " + str(city_with_most_sold[0]))


# Plot a bar chart
#plt.figure(figsize=(10, 6))
group_cities.plot(kind='bar', color='blue')

# Customize the chart
plt.xlabel('Product Combination')
plt.ylabel('Count')
plt.title('Most Common Product Combinations')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

# Display the chart
plt.tight_layout()
plt.show()
```

```python
# MONTH TO SHOW ADVERTISEMENT TO MAXIMISE  PROFIT
# IS THE MONTH WITH HIGHEST REVENUE
import matplotlib.pyplot as plt

edit_monthly_summary=monthly_summary
edit_monthly_summary['Order Date']  = pd.to_datetime(edit_monthly_summary.index.tolist())

print("Month to show advertisement to maximize profit is " + month_with_highest_revenue)
# VISUALIZING THE DATA
plt.plot(edit_monthly_summary["Order Date"], edit_monthly_summary["Price Each"])
plt.xlabel('Order Date')
plt.ylabel('Total Revenue')
plt.title('Customer Purchasing Behavior by Month')
plt.grid(True)
plt.show()
```

```python
# What products are most often sold together?

non_unique_rows = sort_data[sort_data.duplicated(subset='Order ID', keep=False)]


non_unique_product=non_unique_rows.groupby("Order ID").Product.agg(list).reset_index()
product_counts = non_unique_product['Product'].apply(tuple).value_counts()
# Get the most common product combination
most_common_combination = product_counts.idxmax()

print("Products that are most often sold together are", most_common_combination)


```
